```python
df = pd.read_csv("FuelConsumption.csv")

df.head()
```

```python
#summary
df.describe()
```

```python
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head()
```

```python
#visuliaze the data
cdf.hist()
plt.show()
```

```python
#plotting linear relations
plt.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS, color='red')
plt.xlabel('Fuel Consumption COMB')
plt.ylabel('CO2 emissions')
plt.show()
```

```python
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color='purple')
plt.xlabel('Engine Size')
plt.ylabel('CO2 emissions')
plt.show()
```

```python
plt.scatter(cdf.CYLINDERS, cdf.CO2EMISSIONS, color='blue')
plt.xlabel('Cylinders')
plt.ylabel('CO2 emissions')
plt.show()
```

```python
#splitting dataset into train and test sets
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='orange')
plt.xlabel('Engine size')
plt.ylabel('CO2 emmisions')
plt.show()
```

```python
#building model
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit(train_x, train_y)
#the coefficients
print('Coefficients: ',regr.coef_)
print('Intercept: ', regr.intercept_)
```

```python
#plotting predicted line
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='orange')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel('Engine size')
plt.ylabel('CO2 emmisions')
plt.show()
```

```python
from sklearn.metrics import r2_score
test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
test_y_ = regr.predict(test_x)

print('Mean Absolute error: %.2f' % np.mean(np.absolute(test_y_ - test_y)))
print('Residual sum of squares: %.2f' % np.mean((test_y_ - test_y ) ** 2))
print('R-2 score: %.2f' % r2_score(test_y, test_y_))
```
